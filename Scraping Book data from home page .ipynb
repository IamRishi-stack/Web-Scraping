```python
import requests
from tqdm import tqdm
import pandas as pd
from bs4 import BeautifulSoup
```

```python
data = []
for i in tqdm(range(1, 51)):
    link = 'https://books.toscrape.com/catalogue/page-'+ str(i) +'.html'
    
    res = requests.get('https://books.toscrape.com/catalogue/page-2.html')
    soup = BeautifulSoup(res.text, 'html.parser')
    
    #iteratin throughout
    for sp in soup.find_all('li', class_ = 'col-xs-6 col-sm-4 col-md-3 col-lg-3'):
        title        =  sp.find_all('a')[-1].get('title')
        book_link    = 'https://books.toscrape.com/'+sp.find('a').get ('href')
        img_link     = 'https://books.toscrape.com/'+sp.find('img').get ('src')
        rating       =  sp.find('p').get ('class')[1]
        price        =  sp.find('p', class_ = 'price_color').text[2:]
        availibility =  sp.find('p', class_ = 'instock availability').text[15:-6]

        #print('='*30)
        df =  data.append([title,book_link,img_link,rating, price,availibility ])
    
```

```python
df = pd.DataFrame(data, columns = ['title','book_link','img_link','rating','price','availibility'])
```
