```python
import requests 
from tqdm import tqdm
from bs4 import BeautifulSoup
import pandas as pd
```

```python
res = requests.get( 'https://books.toscrape.com/')
sp = BeautifulSoup(res.text, 'html.parser')
```

```python
df = pd.read_csv('book_info.csv')
df.head()
```

```python
data = []

for i in tqdm(df['book_link']):
    
    res = requests.get(i)
    soup = BeautifulSoup(res.text, 'html.parser')
    
    Genre   = soup.find('ul', class_ = 'breadcrumb').find_all('a')[2].text
    UPC     = soup.find('table', class_ = 'table table-striped').find_all('td')[0].text
    Price_x = soup.find('table', class_ = 'table table-striped').find_all('td')[2].text[2:]
    Price_i = soup.find('table', class_ = 'table table-striped').find_all('td')[3].text[2:]
    Tax     = soup.find('table', class_ = 'table table-striped').find_all('td')[4].text[2:]
    Availab = soup.find('table', class_ = 'table table-striped').find_all('td')[5].text
    Review  = soup.find('table', class_ = 'table table-striped').find_all('td')[6].text
    

    data.append([Genre, UPC, Price_x, Price_i,Tax, Availab, Review])
    
```

```python
data
```

```python
df = pd.DataFrame(data, columns = ['typ', 'UPC', 'Price_x', 'Price_i','Tax', 'Availab', 'Review'])
```
